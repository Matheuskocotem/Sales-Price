```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('Data/gas_prices.csv')
```

```python
plt.title('Valores do Combustivel em US$  ')

plt.plot(gas['Year'], gas['Australia'], 'r.--', label = 'Australia')
plt.plot(gas['Year'], gas['Italy'], label = 'Italy')
plt.plot(gas['Year'], gas['USA'], label = 'USA')
plt.plot(gas['Year'], gas['Japan'], label = 'Japan')


plt.xticks(gas['Year'][::2])


plt.xlabel('Years')
plt.ylabel('US$')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_gas_prices(gas: pd.DataFrame, countries: list[str]) -> None:
    """
    Plots gas prices for specified countries over the years.

    Args:
    gas (pd.DataFrame): DataFrame containing gas prices with years as one column and country prices as others.
    countries (list[str]): List of country names to plot.

    Returns:
    None
    """
    for country in countries:
        if country in gas.columns:
            plt.plot(gas['Year'], gas[country], label=country)

    plt.title('Valores do Combustivel em US$')
    plt.xticks(gas['Year'][::2])
    plt.xlabel('Years')
    plt.ylabel('US$')
    plt.legend()
    plt.show()
```
